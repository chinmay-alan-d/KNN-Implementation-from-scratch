```python
mylistX = []
mylisty = []
for i in range(X_train.shape[0]):
    mylistX.append(list(X_train.iloc[i]))
for i in range(y_train.shape[0]):
    mylisty.append(y_train.iloc[i])
```

```python
def knn(trainX,trainY,newinput):
    #trainX=[1.0, 89.0, 66.0, 23.0, 94.0, 28.1, 0.167, 21.0]
    #trainY = 0 or 1
    #test = [1.0, 138.0, 82.0, 0.0, 0.0, 40.1, 0.236, 28.0]
    #sqrt((x2-x1)^2+(y2-y1)^2)
    sumOfsquares = 0
    for i in range(len(trainX)):
        difference = trainX[i] - newinput[i]
        sumOfsquares = sumOfsquares + pow(difference,2)
        
    return math.sqrt(sumOfsquares), trainY
```

```python
def calculateAllDistance(newinput,X_train):
    listOfdistances = []
    for i in range(X_train.shape[0]):
        a = knn(X_train.iloc[i],y_train.iloc[i],newinput)
        listOfdistances.append([a,y_train.iloc[i]])
    listOfdistances.sort()
    return listOfdistances
```

```python
k = math.sqrt(X_train.shape[0])
if k%2 == 0:
    k = k + 1
k
```

```python
def predict(newinput,X_train):
    positives = 0
    negatives = 0
    x = calculateAllDistance(newinput,X_train)
    for i in range(int(k)):
        if(x[i][1] == 1):
            positives += 1
        else:
            negatives += 1
    if positives > negatives:
        return 1
    return 0
```

```python
def verdict(newinput,X_train):
    if predict(newinput,X_train) == 0:
        print('you dont have diabetes!!')
    else:
        print('you have diabetes.')
```

```python
verdict([1.0, 89.0, 76.0, 34.0, 37.0, 31.2, 0.192, 22.0],X_train)
```

```python
correct = 0
wrong = 0
for i in range(X_test.shape[0]):
    l = predict(X_test.iloc[i],X_train)
    if l == y_test.iloc[i]:
        correct += 1
    else:
        wrong += 1
print((correct/X_test.shape[0])*100)
```
